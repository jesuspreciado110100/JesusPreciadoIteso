```python
columnas = ["INDUS", "AGE", "TAX", "RAD", "MEDV"]

datos_filtrados = data[columnas]
datos_filtrados
```

```python
age = datos_filtrados["AGE"]
age = age - min(age)
age = age/max(age)
age.describe()

def scaler(variable) : 
    variable = variable - min(variable)
    variable = variable/max(variable)
    return variable

datos_filtrados.loc[:,"INDUS"] = scaler(datos_filtrados["INDUS"])
datos_filtrados.loc[:,"AGE"] = scaler(datos_filtrados["AGE"])
datos_filtrados.loc[:,"TAX"] = scaler(datos_filtrados["TAX"])
datos_filtrados.loc[:,"RAD"] = scaler(datos_filtrados["RAD"])

datos_filtrados.describe()
```

```python
datos_filtrados = datos_filtrados.dropna()
datos_filtrados.describe()
```
